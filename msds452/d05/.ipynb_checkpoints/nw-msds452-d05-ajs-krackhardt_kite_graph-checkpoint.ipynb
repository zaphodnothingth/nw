```python
#src: https://networkx.org/documentation/stable/auto_examples/algorithms/plot_krackhardt_centrality.html

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import networkx as nx
import pandas as pd
```

```python
G = nx.krackhardt_kite_graph()

b = nx.betweenness_centrality(G)
d = nx.degree_centrality(G)
c = nx.closeness_centrality(G)

df = pd.DataFrame({'bewtweenness': b, 'degree': d, 'closeness': c})
df.style.highlight_max(color = 'lightgreen', axis = 0)
```

```python
colormap = ['white']*3 + ['cyan'] + ['white'] + ['lightgreen']*2 + ['orange'] + ['white']*2
nx.draw(G, with_labels=True, node_color=colormap)
label_b = mpatches.Patch(color='orange', label='betweenness center')
label_c = mpatches.Patch(color='lightgreen', label='closeness center')
label_d = mpatches.Patch(color='cyan', label='highest degree')
plt.legend(handles=[label_b, label_c, label_d], loc='lower left')
plt.show()
```
